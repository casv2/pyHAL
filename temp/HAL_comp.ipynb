```python
import numpy
import pyHAL
from ase.io import read, write
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}'] 
```

```python
al = read("/Users/Cas/Work/HMD/PEG3/peg_all.xyz", ":")
Es1, Fu_s1 = pyHAL.HAL(al[0], "test", nsteps=1000, tau=100, dtau=100.0, ntau=10, Fu_max=0.3, dt=0.5, write_files=False)
```

```python
al = read("/Users/Cas/Work/HMD/PEG3/peg_all.xyz", ":")
Es2, Fu_s2 = pyHAL.HAL(al[0], "test", nsteps=100, tau=0.0, dtau=0.0, ntau=1e10, Fu_max=0.3, dt=0.5, write_files=False)
```

```python
al = read("/Users/Cas/Work/HMD/PEG3/peg_all.xyz", ":")
Es3, Fu_s3 = pyHAL.HAL(al[0], "test", nsteps=1000, tau=100, dtau=25.0, ntau=10, Fu_max=0.3, dt=0.5, write_files=False)
```

```python
dt = 0.5

plt.plot([i*dt for i in range(len(Fu_s1))], Fu_s1, label=r"$\Delta \tau = 100 \, \text{eV}^{-1}$")
plt.plot([i*dt for i in range(len(Fu_s3))], Fu_s3, label=r"$\Delta \tau = 25 \, \text{eV}^{-1}$")
plt.plot([i*dt for i in range(len(Fu_s2))], Fu_s2, label=r"$\Delta \tau = 0 \, \text{eV}^{-1}$")
plt.xlabel("time [fs]")
plt.axhline([0.3], color="black", label=r"$\| F^{\max}_{\sigma^{2}} \|$")
plt.ylabel(r"$ \max \| F_{\sigma^{2}} \|$")
plt.ylim([0.0, 0.35])
plt.legend()
plt.savefig("HAL_tau_comparison.pdf")
#"Δτ = 100 eV⁻¹")
```
